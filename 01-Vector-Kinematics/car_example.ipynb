```python
import sympy as sp
from sympy.physics.mechanics import *
import sympy.physics.mechanics as me
import numpy as np
import matplotlib.pyplot as plt

# Import IPython display for proper LaTeX formatting
from IPython.display import Math

# Initialize symbols
sp.init_printing()

# Enable dot notation printing for dynamicsymbols
from sympy.physics.vector.printing import init_vprinting
init_vprinting(use_latex='mathjax')
```

```python
print("=== DSOLVE EXAMPLE ===")

# Define symbols
t, t1, t2, c = sp.symbols('t t1 t2 c', real=True, positive=True)
C1, C2 = sp.symbols('C1 C2')

# Define position function
x = sp.Function('x')(t)

# First interval: d²x/dt² = 0
eq1 = sp.Eq(sp.diff(x, t, 2), 0)
sol1 = sp.dsolve(eq1)
print("First interval solution:")
display(sol1)

# Second interval: d²x/dt² = -c(t - t1)
eq2 = sp.Eq(sp.diff(x, t, 2), -c*(t - t1))
sol2 = sp.dsolve(eq2)
print("Second interval solution:")
display(sol2)

# Apply initial conditions
sol1_specific = sol1.rhs.subs({C1: 0, C2: 0})
print("With initial conditions x(0)=0, v(0)=0:")
display(sol1_specific)

# Apply continuity conditions
constants = sp.solve([
    sp.Eq(sol2.rhs.subs(t, t1), 0),  # x(t1) = 0
    sp.Eq(sp.diff(sol2.rhs, t).subs(t, t1), 0)  # v(t1) = 0
], [C1, C2])

sol2_specific = sol2.rhs.subs(constants)
print("Second interval with continuity:")
display(sol2_specific)

# Create complete piecewise solution
x_complete = sp.Piecewise(
    (sol1_specific, (t >= 0) & (t < t1)),
    (sol2_specific, (t >= t1) & (t < t2)),
    (sol2_specific.subs(t, t2), t >= t2)
)

print("Complete position function:")
display(Math(r'x(t) = ' + sp.latex(x_complete)))

# Velocity
v_complete = sp.diff(x_complete, t)
print("Complete velocity function:")
display(Math(r'v(t) = ' + sp.latex(v_complete)))

# Acceleration
a_complete = sp.diff(v_complete, t)
print("Complete acceleration function:")
display(Math(r'a(t) = ' + sp.latex(a_complete)))
```
