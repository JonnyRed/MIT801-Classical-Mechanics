# Work and Energy

This notebook covers work, energy, and the work-energy theorem using SymPy's mechanics package.

## Topics Covered:
- Work done by forces
- Kinetic and potential energy
- Work-energy theorem
- Conservation of energy
- Power and efficiency

```python
import sympy as sp
from sympy.physics.mechanics import *
import sympy.physics.mechanics as me
import numpy as np
import matplotlib.pyplot as plt

# Initialize symbols
sp.init_printing()

# Define reference frame
N = ReferenceFrame('N')

# Define symbols
m, g, k = sp.symbols('m g k', positive=True)
t, x, y, z = sp.symbols('t x y z', real=True)
F, v, theta = sp.symbols('F v theta', real=True)
```

## Work Done by a Force

```python
# Define a particle
particle = Particle('particle', mass=m)

# Position and displacement
r = x*N.x + y*N.y + z*N.z
dr = sp.Matrix([sp.diff(x, t), sp.diff(y, t), sp.diff(z, t)])

# Force vector
F_vec = F*sp.cos(theta)*N.x + F*sp.sin(theta)*N.y

# Work done (integral of F · dr)
print("Force vector:")
display(F_vec)

# For constant force, work = F · displacement
dx, dy = sp.symbols('dx dy', real=True)
displacement = dx*N.x + dy*N.y
work = F_vec.dot(displacement)

print(f"\nWork done by constant force: {work}")

# Work done by gravity
W_gravity = m*g*sp.diff(z, t)  # Power (work per unit time)
print(f"\nPower due to gravity: {W_gravity}")
```

## Kinetic Energy

```python
# Define velocity functions
vx = sp.Function('vx')(t)
vy = sp.Function('vy')(t)
vz = sp.Function('vz')(t)

# Velocity vector
v_vec = vx*N.x + vy*N.y + vz*N.z

# Kinetic energy: KE = (1/2)mv²
v_squared = v_vec.dot(v_vec)
KE = sp.Rational(1, 2) * m * v_squared

print("Velocity vector:")
display(v_vec)
print(f"\nSpeed squared: {v_squared}")
print(f"Kinetic energy: {KE}")

# For constant speed v
v_const = sp.symbols('v_const', positive=True)
KE_simple = sp.Rational(1, 2) * m * v_const**2
print(f"\nKinetic energy (constant speed): {KE_simple}")
```

## Potential Energy

```python
# Gravitational potential energy
h = sp.symbols('h', real=True)
PE_gravity = m * g * h

# Elastic potential energy (spring)
x_spring = sp.symbols('x_spring', real=True)
PE_spring = sp.Rational(1, 2) * k * x_spring**2

print("Potential energy:")
print(f"Gravitational: {PE_gravity}")
print(f"Elastic (spring): {PE_spring}")

# Force from potential energy (negative gradient)
F_gravity = -sp.diff(PE_gravity, h)
F_spring = -sp.diff(PE_spring, x_spring)

print(f"\nForce from potential:")
print(f"Gravitational force: {F_gravity}")
print(f"Spring force: {F_spring}")
```

## Work-Energy Theorem

```python
# Work-energy theorem: W_net = ΔKE

# Initial and final velocities
v_i, v_f = sp.symbols('v_i v_f', positive=True)

# Initial and final kinetic energy
KE_i = sp.Rational(1, 2) * m * v_i**2
KE_f = sp.Rational(1, 2) * m * v_f**2

# Change in kinetic energy
delta_KE = KE_f - KE_i

# Work done equals change in kinetic energy
work_done = delta_KE

print("Work-Energy Theorem:")
print(f"Initial KE: {KE_i}")
print(f"Final KE: {KE_f}")
print(f"ΔKE: {delta_KE}")
print(f"Work done: {work_done}")

# Solve for final velocity given work
W_given = sp.symbols('W_given', real=True)
v_f_solution = sp.solve(sp.Eq(W_given, delta_KE), v_f)
print(f"\nFinal velocity given work W: {v_f_solution}")
```

## Conservation of Energy

```python
# Total mechanical energy
E_total = KE + PE_gravity + PE_spring

print("Total mechanical energy:")
print(f"E_total = KE + PE_gravity + PE_spring")
print(f"E_total = {KE} + {PE_gravity} + {PE_spring}")

# For a falling object (no air resistance)
# Energy is conserved: KE_i + PE_i = KE_f + PE_f
h_i, h_f = sp.symbols('h_i h_f', real=True)

PE_i = m * g * h_i
PE_f = m * g * h_f

conservation_eq = sp.Eq(KE_i + PE_i, KE_f + PE_f)
print(f"\nConservation equation: {conservation_eq}")

# Solve for final velocity
v_f_from_height = sp.solve(conservation_eq, v_f)
print(f"\nFinal velocity from height: {v_f_from_height}")

# Simplify for case where initial velocity is zero
v_f_zero_initial = [sol.subs(v_i, 0) for sol in v_f_from_height]
print(f"\nFinal velocity from rest: {v_f_zero_initial}")
```

## Example: Pendulum Energy

```python
# Pendulum parameters
L, theta_max = sp.symbols('L theta_max', positive=True)
theta_pend = sp.Function('theta')(t)

# Height of pendulum bob
h_pend = L * (1 - sp.cos(theta_pend))

# Velocity of pendulum bob
omega = sp.diff(theta_pend, t)  # Angular velocity
v_pend = L * omega  # Linear velocity

# Energies
KE_pend = sp.Rational(1, 2) * m * v_pend**2
PE_pend = m * g * h_pend
E_pend = KE_pend + PE_pend

print("Pendulum energy:")
print(f"Height: {h_pend}")
print(f"Velocity: {v_pend}")
print(f"Kinetic energy: {KE_pend}")
print(f"Potential energy: {PE_pend}")
print(f"Total energy: {E_pend}")

# At maximum angle (velocity = 0)
E_max = PE_pend.subs(theta_pend, theta_max)
print(f"\nTotal energy at maximum angle: {E_max}")

# Find velocity at bottom (θ = 0)
E_bottom = KE_pend.subs(theta_pend, 0) + PE_pend.subs(theta_pend, 0)
energy_conservation = sp.Eq(E_max, E_bottom)
omega_bottom = sp.solve(energy_conservation, omega)[0]
v_bottom = L * omega_bottom

print(f"\nVelocity at bottom: {v_bottom}")
```

## Power

```python
# Power: P = F · v = dW/dt

# Power for constant force and velocity
P_const = F * v_const

# Power as time derivative of work
W_function = sp.Function('W')(t)
P_derivative = sp.diff(W_function, t)

print("Power:")
print(f"Constant force and velocity: P = {P_const}")
print(f"As derivative: P = dW/dt = {P_derivative}")

# Power for falling object
v_falling = sp.diff(h, t)  # dh/dt = v
P_gravity = m * g * v_falling  # Negative when falling
print(f"\nPower due to gravity: {P_gravity}")

# Average power
W_total, t_total = sp.symbols('W_total t_total', positive=True)
P_avg = W_total / t_total
print(f"\nAverage power: {P_avg}")
```

## Energy in Simple Harmonic Motion

```python
# Simple harmonic motion: x = A*cos(ωt)
A, omega_shm = sp.symbols('A omega_shm', positive=True)
x_shm = A * sp.cos(omega_shm * t)
v_shm = sp.diff(x_shm, t)

# Energy in SHM
KE_shm = sp.Rational(1, 2) * m * v_shm**2
PE_shm = sp.Rational(1, 2) * k * x_shm**2
E_shm = KE_shm + PE_shm

print("Simple harmonic motion energy:")
print(f"Position: {x_shm}")
print(f"Velocity: {v_shm}")
print(f"Kinetic energy: {KE_shm}")
print(f"Potential energy: {PE_shm}")

# Simplify using ω² = k/m
E_shm_simplified = sp.simplify(E_shm.subs(k, m*omega_shm**2))
print(f"\nTotal energy (simplified): {E_shm_simplified}")

# Maximum energy (at turning points)
E_max_shm = sp.Rational(1, 2) * k * A**2
print(f"\nMaximum energy: {E_max_shm}")
```

## Numerical Examples

```python
# Example: Object falling from height
values = {m: 2, g: 9.81, h_i: 10, h_f: 0, v_i: 0}

# Final velocity from energy conservation
v_f_num = float(v_f_zero_initial[0].subs(values))
KE_final_num = float(KE_f.subs({m: 2, v_f: v_f_num}))
PE_initial_num = float(PE_i.subs(values))

print("Numerical example - falling object:")
print(f"Mass: 2 kg, Height: 10 m")
print(f"Final velocity: {v_f_num:.2f} m/s")
print(f"Final kinetic energy: {KE_final_num:.2f} J")
print(f"Initial potential energy: {PE_initial_num:.2f} J")
print(f"Energy conservation check: {abs(KE_final_num - PE_initial_num) < 0.01}")

# Example: Spring-mass system
values_spring = {m: 1, k: 100, A: 0.1}
E_max_num = float(E_max_shm.subs(values_spring))
omega_num = float(omega_shm.subs({k: 100, m: 1}))

print(f"\nSpring-mass system:")
print(f"Mass: 1 kg, Spring constant: 100 N/m, Amplitude: 0.1 m")
print(f"Maximum energy: {E_max_num:.2f} J")
print(f"Natural frequency: {omega_num:.2f} rad/s")
```
