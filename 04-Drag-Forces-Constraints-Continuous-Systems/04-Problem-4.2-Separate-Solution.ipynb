# Problem 4.2 Separate Solution

A particle of mass $m$ moves along the $x$-axis under a velocity-dependent drag force
$$\vec{F} = -b e^{cv} \hat{i},$$
with $b>0$, $c>0$, and initial condition $v(0)=v_0$. Find $v(t)$.

## Derivation

Using Newton's second law in 1D:
$$m \frac{dv}{dt} = -b e^{cv}. $$
Separate variables:
$$e^{-cv}  dv = -\frac{b}{m}  dt. $$
Integrate both sides:
$$-\frac{1}{c}e^{-cv} = -\frac{b}{m}t + C. $$
Apply $v(0)=v_0$:
$$-\frac{1}{c}e^{-cv_0} = C. $$
So
$$e^{-cv} = e^{-cv_0} + \frac{bc}{m}t. $$
Take the logarithm:
$$v(t) = -\frac{1}{c}\ln\left(e^{-cv_0}+\frac{bc}{m}t\right).$$

## Final Answer

$$\boxed{v(t) = -\frac{1}{c}\ln\left(e^{-cv_0}+\frac{bc}{m}t\right)}$$

```python
import sympy as sp
from IPython.display import display, Math

m, b, c, v0, t = sp.symbols("m b c v0 t", positive=True, real=True)
v_expr = -sp.log(sp.exp(-c * v0) + b * c * t / m) / c

dv_dt = sp.diff(v_expr, t)
ode_residual = sp.simplify(m * dv_dt + b * sp.exp(c * v_expr))
ic_residual = sp.simplify(v_expr.subs(t, 0) - v0)

display(Math(r"v(t)=" + sp.latex(v_expr)))
display(Math(r"m\,\dot v + b e^{cv} = " + sp.latex(ode_residual)))
display(Math(r"v(0)-v_0 = " + sp.latex(ic_residual)))
```
