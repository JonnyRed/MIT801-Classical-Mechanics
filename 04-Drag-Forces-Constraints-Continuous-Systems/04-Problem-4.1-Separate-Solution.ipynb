# Problem 4.1 - Separate Symbolic Solution

This notebook solves parts (a), (b), and (c) for the block-pulley system using SymPy.

```python
import sympy as sp
from sympy.physics.mechanics import ReferenceFrame, dynamicsymbols
import sympy.physics.vector as spv
from IPython.display import display, Math, Markdown

sp.init_printing()

m1, m2, mu_s, mu_k, theta, g = sp.symbols(
    "m1 m2 mu_s mu_k theta g", positive=True, real=True
)
```

## (a) and (b) Static thresholds

Choose +x up the incline for block 1. For a massless, frictionless pulley/string, the hanging block sees upward force $2T$, so at static threshold $m_2 g = 2T$.

```python
# (a) block 1 just barely slides up -> static friction acts down the incline
T_up = sp.simplify(m1 * g * (sp.sin(theta) + mu_s * sp.cos(theta)))
m2_min = sp.simplify(2 * T_up / g)

# (b) block 1 just barely slides down -> static friction acts up the incline
T_down = sp.simplify(m1 * g * (sp.sin(theta) - mu_s * sp.cos(theta)))
m2_max = sp.simplify(2 * T_down / g)

display(Math(r"m_{2,\min} = " + sp.latex(m2_min)))
display(Math(r"m_{2,\max} = " + sp.latex(m2_max)))
```

## (c) Sliding upward with kinetic friction

Assume block 1 moves up the incline, so kinetic friction on block 1 points down the incline.

Use:
- Block 1: $T - m_1 g\sin\theta - \mu_k m_1 g\cos\theta = m_1 a_1$
- Hanging mass (downward positive): $m_2 g - 2T = m_2 a_2$
- Rope constraint magnitude: $a_1 = 2 a_2$.

```python
a1, a2, T = sp.symbols("a1 a2 T", real=True)

eq1 = sp.Eq(T - m1 * g * sp.sin(theta) - mu_k * m1 * g * sp.cos(theta), m1 * a1)
eq2 = sp.Eq(m2 * g - 2 * T, m2 * a2)
eq3 = sp.Eq(a1, 2 * a2)

sol = sp.solve([eq1, eq2, eq3], [a1, a2, T], dict=True)[0]
a1_expr = sp.simplify(sol[a1])

display(Math(r"a_{x1} = " + sp.latex(a1_expr)))
display(Math(r"|a_{x1}| = " + sp.latex(sp.Abs(a1_expr))))
display(
    Math(
        r"\text{Upward-motion consistency condition: }" 
        r" m_2 > 2 m_1(\sin\theta + \mu_k \cos\theta)"
    )
)
```
